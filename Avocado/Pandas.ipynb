```python
import pandas as pd

df = pd.read_csv('datasets/avocado.csv')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
albany = df[df['region'] == 'Albany']
albany.head()
```

```python
albany.tail()
```

```python
albany.set_index('Date')
```

```python
albany = albany.set_index('Date')
# same as albany.set_index('Date', inplace=true)
```

```python
albany.head()
```

```python
import matplotlib.pyplot as plt
albany.plot()
```

```python
albany['AveragePrice'].plot()
```

```python
# graph looks really busy
# rolling 25 records then get the mean.
albany['AveragePrice'].rolling(25).mean().plot()
```

```python
# makes dates in proper order
albany.sort_index(inplace=True)
albany['AveragePrice'].rolling(25).mean().plot()
```

```python
# create new column named avgprice25ma.
albany['avgprice25ma'] = albany['AveragePrice'].rolling(25).mean()
albany.head()
```
